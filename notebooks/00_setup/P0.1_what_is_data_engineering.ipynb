## Key Concepts

**Data** — Raw facts. Example: "Dhiru bought shoes for ₹2000 at 3pm"

**Database** — Organised place to store data so you can find it quickly

**Data Pipeline** — The journey data takes from where it is created to where it becomes useful

**OLTP** — Online Transaction Processing. One transaction at a time. Fast. Current data. Example: cashier billing one customer

**OLAP** — Online Analytical Processing. Millions of records. Analysis. Historical data. Example: HQ manager viewing full year sales

**ETL** — Extract → Transform → Load. Old way. Clean data first, then send.

**ELT** — Extract → Load → Transform. Modern way. Send raw data first, clean later in Databricks.

## Real World Analogy — Big Bazaar

| Concept | Big Bazaar Example |
|---------|-------------------|
| Data | All products in the store |
| Database | Store shelves — organised by category |
| Pipeline | Supply chain — farm to truck to shelf to customer |
| OLTP | Cashier billing one customer at the counter |
| OLAP | HQ manager viewing sales across all 50 stores |
| ELT | Raw stock arrives at warehouse first, sorted later |

**Big Bazaar Store = OLTP**  
**Big Bazaar HQ = OLAP**

## How This Applies to RetailPulse Project

Our project follows ELT method with Medallion Architecture:

E = Extract    — Pull raw Olist data from source
L = Load       — Land raw data into BRONZE layer (untouched)
T = Transform  — Clean and validate in SILVER layer
— Business aggregates in GOLD layer

**Bronze** = Load step — raw data arrives exactly as it came  
**Silver** = Transform step — cleaning, validation, fixing  
**Gold**   = Serving step — ready for manager/dashboard  

Databricks sits in the middle and does all the ELT work.

## Checkpoint Q&A

**Q1 — What is the difference between OLTP and OLAP?**  
OLTP = One transaction at a time, fast, current data (store cashier)  
OLAP = Millions of records, analysis, historical data (HQ manager)

**Q2 — What does ELT stand for?**  
E = Extract — take data from source  
L = Load — put raw data into Bronze first  
T = Transform — clean data in Silver layer  

**Q3 — Which RetailPulse layer is Load and which is Transform?**  
Bronze = Load step  
Silver = Transform step

## Key Takeaways

1. Data Engineering exists to move data from where it is created to where it is useful
2. OLTP handles transactions — OLAP handles analysis — never mix them
3. ELT is the modern way — Databricks is powerful enough to transform after loading
4. Bronze = raw, Silver = clean, Gold = business ready
5. Every pipeline follows Extract → Load → Transform in our project
